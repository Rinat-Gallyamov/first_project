Определить цикл сделки(время, которое проходит с момента первого касания с клиентом до момента получения денег)
1. Отбираем id людей которые оплатили заказ
2. Находим id  и дату регистрации тех, что есть в data2(Зарегестрированные)
3. Считаем среднее в разнице между даты оплаты и даты регистрации. 
Примерно 11 дней

```python
id_money = data3['ID пользователя'][(data3['Статус'] == 'Завершен') & (pd.isnull(data3['Дата оплаты']) == False)]
id_money.value_counts()
```

```python
data2['id'] = data2['id'].apply(str)
data2['id']
```

```python
reg_id_money = data2['id'][data2['id'].isin(id_money)]
reg_id_money
## Казалось бы количество людей здесь должно быть равно количеству людей сделавших заказ,
## а именно 6105. Видимо данные в data2 не полные (возможно отсутствуют старые клиенты)
```

```python
time_reg = data2[data2['id'].isin(reg_id_money)]
time_reg = time_reg.iloc[:,[0,3]]
time_reg
```

```python
time = data3[(data3['ID пользователя'].isin(time_reg['id'])) & (data3['Статус'] == 'Завершен')]
time = time.iloc[:,[2,7]]
time['Дата оплаты'] = pd.to_datetime(time['Дата оплаты'])
time = time.groupby(['ID пользователя'],sort=False).agg('min')
time.sort_index(ascending=False,inplace=True,axis=1)
time
```

```python
time_reg['Создан'] = pd.to_datetime(time_reg['Создан'])
time_reg = time_reg.groupby(['id'],sort=False).agg('min')
time_reg.sort_index(ascending=False,inplace=True,axis=1)
time_reg
```

```python

((time['Дата оплаты']-time_reg['Создан']).dt.days).mean()
```
